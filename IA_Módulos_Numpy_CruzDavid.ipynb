Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con:

```python
def resumen(lista):
    # Validamos que la lista no esté vacía para evitar error al calcular promedio
    if not lista:
        return "La lista está vacía"

    # Calculamos los valores
    longitud = len(lista)
    suma = sum(lista)
    promedio = suma / longitud
    minimo = min(lista)
    maximo = max(lista)

    # Retornamos el diccionario
    return {
        'longitud': longitud,
        'suma': suma,
        'promedio': promedio,
        'minimo': minimo,
        'maximo': maximo
    }

# Ejemplo de uso:
mi_lista = [10, 20, 30, 40, 50]
resultado = resumen(mi_lista)
print(resultado)
```

Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].

```python
# 1. Definimos la función
def potencia(x, n=3):
    return x ** n

numeros = [1, 2, 3, 4, 5]

# 2. Aplicamos map
resultado = list(map(potencia, numeros))

print(resultado)
# Resultado: [1, 8, 27, 64, 125]
```

Ejercicio 4: Crea una matriz identidad 5×5 y verifica que la suma de su diagonal
  es igual a 5 (usa np.trace()).

```python
import numpy as np

# 1. Creamos la matriz identidad 5x5
# np.eye(5) genera una matriz con 1s en la diagonal y 0s en el resto
matriz_id = np.eye(5)

# 2. Calculamos la traza (suma de la diagonal principal)
suma_diagonal = np.trace(matriz_id)

# 3. Verificamos los resultados
print("Matriz Identidad 5x5:")
print(matriz_id)
print("-" * 30)
print(f"Suma de la diagonal (np.trace): {suma_diagonal}")

# Verificación lógica
if suma_diagonal == 5:
    print("¡Verificación exitosa! La suma es igual a 5.")
```

Ejercicio 4: Demuestra la diferencia vista vs. copia: modifica un elemento
  en un slice y verifica que el array original también cambia.

```python
import numpy as np

# 1. Creamos el array original
original = np.array([10, 20, 30, 40, 50])
print(f"Array original: {original}")

# 2. Creamos una VISTA usando slicing
vista = original[0:3]
print(f"Vista (primeros 3 elementos): {vista}")

# 3. MODIFICAMOS la vista
vista[0] = 99
print("\n--- Después de modificar vista[0] = 99 ---")

# 4. Verificamos ambos
print(f"Vista modificada: {vista}")
print(f"Array original:   {original}  <-- ¡CAMBIÓ!")

# 5. Demostración de COPIA real
copia_real = original.copy()
copia_real[1] = 777
print("\n--- Después de modificar copia_real[1] = 777 ---")
print(f"Copia modificada: {copia_real}")
print(f"Array original:   {original}  <-- ¡NO cambió!")
```

Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).


```python
np.random.seed(7)
matriz = np.random.randint(1, 101, size=(5, 4))
print("Matriz Original:")
print(matriz)
print("-" * 40)
```

  a) Calcula la media, máximo y mínimo de cada columna.


```python
medias = matriz.mean(axis=0)
maximos = matriz.max(axis=0)
minimos = matriz.min(axis=0)
print("Estadísticas por columna:")
print(f"Medias:  {medias}")
print(f"Máximos: {maximos}")
print(f"Mínimos: {minimos}")
```

 b) Calcula la suma acumulada de la primera fila.


```python
suma_acum_fila1 = np.cumsum(matriz[0, :])
print(f"\nSuma acumulada de la fila 1: {suma_acum_fila1}")
```

 c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
     Fórmula: (x - min_col) / (max_col - min_col)

```python
matriz_norm = (matriz - minimos) / (maximos - minimos)

print("\nMatriz Normalizada (valores entre 0 y 1):")
print(matriz_norm.round(4))
```

Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.
  Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
import numpy as np
# Usamos arange para que los números sean distintos y sea fácil ver la unión
A = np.arange(1, 13).reshape(3, 4)
B = np.arange(13, 25).reshape(3, 4)

print("Matriz A:\n", A)
print("Matriz B:\n", B)
print("-" * 30)

# Resultado esperado: (3, 8)
concat_h = np.hstack((A, B))
print("Concatenación Horizontal (hstack):")
print(concat_h)

# Resultado esperado: (6, 4)
concat_v = np.vstack((A, B))
print("\nConcatenación Vertical (vstack):")
print(concat_v)

# Como tiene 6 filas, cada parte tendrá 2 filas
partes = np.vsplit(concat_v, 3)

print("\n--- Resultado de vsplit (3 partes) ---")
for i, p in enumerate(partes):
    print(f"Parte {i+1} (shape {p.shape}):")
    print(p)
```
